```python
import sys
sys.path.append('..')

import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from models.LinearRegression import LinearRegression 


SEED = 42

X, y = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=SEED)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED)

rg = LinearRegression(lr=0.01)
rg.fit(X_train, y_train)

y_pred = rg.predict(X_test)
mse = rg.mean_squared_error(y_test, y_pred)
print(mse)
y_pred_line = rg.predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X, y)
plt.plot(X, y_pred_line, color='red')
plt.show()
```
